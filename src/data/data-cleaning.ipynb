# Data cleaning

```python
import pandas as pd
```

```python
global_population_1970_2022_df = pd.read_csv("unclean-data/global_population-1970-2022.csv")

global_population_1970_2022_df.head()
```

```python
global_population_2020_df = pd.read_csv("unclean-data/global_population-2020.csv")

global_population_2020_df.head()
```

## Merging datasets

```python
global_population_1970_2022_df = global_population_1970_2022_df.rename(columns={"Country/Territory": "Country"})
global_population_2020_df = global_population_2020_df.rename(columns={"Country (or dependency)": "Country"})
```

```python
global_population_df = pd.merge(global_population_1970_2022_df, global_population_2020_df, on="Country", how="left")

global_population_df.head()
```

## Columns

```python
global_population_df["Population (2020)"].head()
```

```python
global_population_df.columns
```

### Dropping columns

```python
global_population_df = global_population_df.drop(columns=["CCA3", "Capital", "Continent", "2022 Population", "2015 Population", "Area (km²)", "Density (per km²)", "Growth Rate", "Population (2020)", "Density (P/Km²)", "Land Area (Km²)", "World Share"])

global_population_df.head()
```

### Reordering columns

```python
global_population_df = global_population_df[["Rank", "Country", "Med. Age", "Urban Pop %", "Fert. Rate", "Migrants (net)", "1970 Population", "1980 Population", "1990 Population", "2000 Population", "2010 Population", "2020 Population", "Net Change", "Yearly Change", "World Population Percentage"]]

global_population_df.head()
```

### Renaming columns

```python
global_population_df = global_population_df.rename(columns={
    "Med. Age": "Median Age",
    "Urban Pop %": "Percent Urban",
    "Fert. Rate": "Fertility Rate",
    "Migrants (net)": "Net Migrants",
    "World Population Percentage": "Global Population (%)"
})

global_population_df.head()
```

### Fixing types

```python
global_population_df.info()
```

#### Fixing types: numerical

```python
global_population_df["Median Age"] = global_population_df["Median Age"].replace("N.A.", None).astype("Int64")

global_population_df["Percent Urban"] = global_population_df["Percent Urban"].apply(lambda percent: str(percent).replace(" %", ""))
global_population_df["Percent Urban"] = global_population_df["Percent Urban"].replace({"N.A.":None, "nan":None}).astype("Int64")

global_population_df["Fertility Rate"] = global_population_df["Fertility Rate"].replace("N.A.", None).astype("float")

global_population_df["Net Migrants"] = global_population_df["Net Migrants"].astype("Int64")

global_population_df["Net Change"] = global_population_df["Net Change"].astype("Int64")

global_population_df["Yearly Change"] = global_population_df["Yearly Change"].apply(lambda change: str(change).replace(" %", ""))
global_population_df["Yearly Change"] = global_population_df["Yearly Change"].astype("float")

global_population_df.info()
```

### Handling null values

### Handling null values: numerical

```python
global_population_df = global_population_df.dropna(subset=["Median Age", "Percent Urban", "Fertility Rate", "Net Migrants", "Net Change", "Yearly Change"])

global_population_df.info()
```
